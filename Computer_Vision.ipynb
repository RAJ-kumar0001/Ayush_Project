```python
model = YOLO('yolov8s.pt')
```

```python
results = model.train(data='/content/drive/MyDrive/aquarium_pretrain/data.yaml', epochs=100, imgsz=640)
```

```python
metrics = model.val()
```

```python
def detection_pipline(images):
    output = model(images)
    
    # Visualize the results
    for i, r in enumerate(output):
        # Plot results image
        im_bgr = r.plot()  # BGR-order numpy array
        im_rgb = Image.fromarray(im_bgr[..., ::-1])  # RGB-order PIL image

        plt.axis('off')
        plt.imshow(im_rgb)
        plt.show()
```
